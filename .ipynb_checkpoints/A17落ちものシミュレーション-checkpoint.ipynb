```python
def main():
    first_line = input()
    H, W, N = first_line.split(' ')
    
    simulation_data = []
    
    # init data
    for row in range(int(H)):
        row_data = []
        for col in range(int(W)):
            row_data.append('.')
        simulation_data.append(row_data)
        
    for index in range(int(N)):
        input_line = input()
        h,w,x = input_line.split(' ')
        simulation_data = update_simulation_data(simulation_data, int(h), int(w), int(x))
    
    print_simulation(simulation_data)

def update_simulation_data(simulation_data, h, w, x):
    update_data = simulation_data.copy()
    
    for i in range(len(update_data)):
#         row = update_data[len(update_data) - 1 - i]
#         slice_row = row[x:x+w]
#         if slice_row.count('.') == w:
#             #insert data
        if is_ok_to_insert(update_data, i, h, w, x):
            update_data = insert_block(update_data, i, h, w, x)
#             print_simulation(update_data)
            break
            
    return update_data

def is_ok_to_insert(data, row_number, h, w, x):
    data_cp = data.copy()
    
    count_height = 0
    for i in range(len(data_cp)  - row_number):
        row = data[len(data_cp) - 1 - i - row_number]
        slice_row = row[x:x+w]
        if slice_row.count('.') == w:
            count_height += 1
#     print(count_height)
    
    if count_height == len(data_cp)  - row_number:
        return True
    else:
        return False
            
def insert_block(data, row_number, h, w, x):
    data_cp = data.copy()
    
    for i in range(h):
        update_row = data_cp[len(data) - 1 - i -  row_number]
        for j in range(w):
            update_row[x + j] = '#'
    
    return data_cp

def print_simulation(simulation_data):
    for i in range(len(simulation_data)):
        row = simulation_data[i]
        str_row = ''
        for index in range(len(row)):
            str_row += row[index]
        print(str_row)
main()
```
