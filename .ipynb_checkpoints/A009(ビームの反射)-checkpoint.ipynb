```python
# coding: utf-8
# 自分の得意な言語で
# Let's チャレンジ！！

STRAIGHT_FORWARD = 0
STRAIGHT_BACKWARD = 1
STRAIGHT_UP = 2
STRAIGHT_DOWN = 3


def create_representive_data():
    input_HW = input()
    W, H = input_HW.split(' ')

    representive_data = []

    if H.isdigit() and W.isdigit():
        for w in range(int(W)):
            one_line_arr_data = []
            input_line = input()
            for h in range(int(H)):
                one_line_arr_data.append(input_line[h])
            representive_data.append(one_line_arr_data)

    return representive_data


def get_direction(previous_direction, beam):
    if beam == '_':
        direction = previous_direction
    elif beam == '\\':
        if previous_direction == STRAIGHT_FORWARD:
            direction = STRAIGHT_DOWN
        elif previous_direction == STRAIGHT_BACKWARD:
            direction = STRAIGHT_UP
        elif previous_direction == STRAIGHT_UP:
            direction = STRAIGHT_BACKWARD
        else:
            direction = STRAIGHT_FORWARD
    else:
        if previous_direction == STRAIGHT_FORWARD:
            direction = STRAIGHT_UP
        elif previous_direction == STRAIGHT_BACKWARD:
            direction = STRAIGHT_DOWN
        elif previous_direction == STRAIGHT_UP:
            direction = STRAIGHT_FORWARD
        else:
            direction = STRAIGHT_BACKWARD

    return direction


def get_position_box_now(previous_h, previous_w, direction):
    if direction == STRAIGHT_FORWARD:
        h = previous_h
        w = previous_w + 1
    elif direction == STRAIGHT_BACKWARD:
        h = previous_h
        w = previous_w - 1
    elif direction == STRAIGHT_UP:
        h = previous_h - 1
        w = previous_w
    else:
        h = previous_h +1
        w = previous_w
    return h, w


def count_through_boxes():
    beam_data = create_representive_data()
    h = 0
    w = 0
    W = len(beam_data[0])
    H = len(beam_data)
    direction = STRAIGHT_FORWARD
    through_boxes = 0
    while 1:
        direction = get_direction(direction, beam_data[h][w])
        h, w = get_position_box_now(h, w, direction)
        through_boxes += 1
        if h >= H or w >= W or h < 0 or w < 0:
            break

    print(through_boxes)
    return through_boxes


count_through_boxes()
```
